```python
from types import new_class
# นำเข้าไลบรารี
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

#โหลดชุดข้อมูล Iris Dataset
iris = datasets.load_iris()
X = iris.data #Features matrix
y = iris.target # Labels (0 = Setosa, Versicolor, Virginiga)

#แปลงเป็น DataFrame เพื่อดูข้อมูล
df = pd.DataFrame(X, columns=iris.feature_names)
df['target'] = y #เพิ่มคอลัมน์ target เพื่อใช้เปรียบเทียบ

#ใช้ K-Means Clustering แบ่งข้อมูลเป็น 3 กลุ่ม
kmeans = KMeans(n_clusters=3, random_state=42,n_init=10)
df['cluster'] = kmeans.fit_predict(X)

# แสดงค่าเฉลี่ยของแต่ละคลัสเตอร์
cluster_means = df.groupby('cluster').mean()
print("ค่าเฉลี่ยของแต่ละคลัสเตอร์:")
display(cluster_means) # ใช้ display() เพื่อแสดง DataFrame ใน Colab

# ใช้ PCA เพื่อลดมิติข้อมูลจาก 4 มิติ  → 2 มิติ สำหรับการแสดงผล
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# แสดงผลลัพธ์ของการจัดกลุ่มเป็นกราฟ
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=df['cluster'], cmap='viridis',edgecolor='k')
plt.xlabel('PCA  1')
plt.ylabel('PCA  2')
plt.colorbar(label='Cluster')
plt.show()

# ทดสอบข้อมูลใหม่
new_data = np.array([[5.5,2.6,4.4,1.2], # น่าจะอยู่ใกล้ Versicolor
                     [6.8,3.2,5.9,2.3]]) # น่าจะอยู่ใกล้ Virginiga
# ทำนายกลุ่มของข้อมูลใหม่
predicted_clusters = kmeans.predict(new_data)
print("กลุ่มของข้อมูลใหม่:")
print(predicted_clusters)
```

```python
# ตรวจสอบว่าแต่ละคลัสเตอร์หมายถึงดอกไม้ชนิดใด
df.groupby('cluster')['target'].value_counts()

# Cluster 0 → Setosa
# Cluster 1 → Versicolor
# Cluster 2 → Virginiga
```

```python
# ลดมิติของข้อมูลใหม่ด้วย PCA
new_data_pca = pca.transform(new_data)

# วาดการฟ K-Means clustering
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=df['cluster'], cmap='viridis', edgecolor='k')
plt.scatter(new_data_pca[:, 0], new_data_pca[:, 1], c='red', marker='x', label='New Data')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.colorbar(label='Cluster')
plt.legend()
plt.show()
```
